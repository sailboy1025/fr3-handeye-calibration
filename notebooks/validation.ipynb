````python
import numpy as np
from scipy.spatial.transform import Rotation as R
from spatialmath import SE3
def pose_to_matrix(px, py, pz, qx, qy, qz, qw):
    T = np.eye(4)
    T[:3, :3] = R.from_quat([qx, qy, qz, qw]).as_matrix()
    T[:3, 3] = [px, py, pz]
    return T
````

````python
tag_T_ee = SE3().Tz(0.06) @ SE3().Rz(90, unit='deg') @ SE3.Ry(-90, unit='deg')
````

````python
ee_T_base = pose_to_matrix(
    0.5984203219413757,
    0.3618444502353668,
    0.20653866231441498,
    -0.22004627779213634,
    0.6877651406407475,
    0.6383786905712042,
    0.2665171558715004,
)
tag_T_base = ee_T_base @ tag_T_ee.A
cam_T_base = np.array([
    [ 0.0241404614906644, -0.44315264823165645,  0.8961210679837226, -0.09965914957225427],
    [-0.9993724798166944, -0.03393909490325564,  0.010138265245033196, 0.47240773540515774],
    [ 0.02592073887929497,-0.8958034763286074,  -0.4436938664141459, 0.4392741508201531],
    [0,0,0,1]
])

tag_T_cam = np.linalg.inv(cam_T_base) @ tag_T_base

print(tag_T_cam)
````
